## netlify uses `index.html`

An index page at the root of this repo is needed to display something at the root of <https://rasx-node-js.netlify.app/>. My plan today is to:

- define “index categories” of study by eyeballing the directories in this repo
- generate a directory listing grouped by the categories established above
- save these data to a JSON file that can be loaded in the root `index.html` file

```python
from os import path, walk
from datetime import datetime
from pathlib import Path
import json
```

## index categories

Instead of showing a gigantic list of my samples, I have broken them down by category, based on the conventional prefix of sample directory names:

```python
index_categories = {
    'alpine-js': 'Nothing here yet. New material is on the way.',
    'angular': 'The latest official Angular walkthroughs.',
    'astro': 'Nothing here yet. New material is on the way.',
    'bulma': 'Bulma CSS samples for Studio-wide conventions.',
    'css': 'Largely historical CSS experiments and samples.',
    'd3': 'D3.js experiments and walkthroughs.',
    'eleventy': 'Eleventy (a.k.a. Build Awesome) experiments and samples.',
    'elm': 'Elm experiments.',
    'html5': 'Experiments and samples for plain vanilla HTML5.',
    'jquery': 'Experiments and samples for jQuery.',
    'karma': 'Samples based on tutorial material for Karma.',
    'mdn': 'Experiments and samples based on material from MDN (Mozilla Developer Network).',
    'mermaid': 'Samples based on tutorial material for Mermaid.',
    'mocha': 'Samples based on researched material for Mocha.',
    'node': 'Experiments and samples for Node.js.',
    'npm': 'Experiments and samples for Node Package Manager.',
    'react': 'Samples based on tutorial material for React.',
    'svg': 'Experiments and samples for SVG.',
    'tiffany-rayside': 'Samples based on material from Tiffany Rayside.',
    'typescript': 'Experiments and samples based on tutorial material for Typescript.',
    'vite': 'Nothing here yet. New material is on the way.',
    'web-components': 'Experiments and samples based on tutorial material for Web Components.',
    'webpack': 'Samples based on tutorial material for webpack.'
}
```

## walking through the Node.js sample directories

When we search through all the top-level directories, we will do this for every key in the `index_categories` dictionary. For every key, we will:

1. look for an `index.html` file in the top-level directory
2. look for an `index.html` file in the child directories of the top-level directory

### the `get_documents` function

We will look for these `index.html` files with the `get_documents` function below:

```python
walk_root = '../../'
walked = list(walk(walk_root))

def get_documents(dir_prefix: str):

    def list_comprehension_predicate(dir_path, dir_names, file_names):
        filtered_by_prefix = dir_path.find(f'{walk_root}{dir_prefix}') != -1
        has_index_file = [name for name in file_names].count('index.html') > 0
        no_node_folders = dir_path.find('node_modules') == -1
        no_src_folders = dir_path.find('src') == -1
    
        return filtered_by_prefix and has_index_file and no_node_folders and no_src_folders

    directories = [
        (f'{dir_path.replace(walk_root, '')}/index.html')
            for (dir_path, dir_names, file_names)
            in walked
            if list_comprehension_predicate(dir_path, dir_names, file_names)
    ]

    return sorted(directories)
```

### generating the JSON

The paths to these `index.html` files will be saved in the `json_object` dictionary below:

```python
key_for_documents = 'documents'

json_object = {
    'inceptDate': datetime.now().isoformat(),
    'segmentName': 'Node.js samples and experiments',
    key_for_documents: []
}

for k, v in index_categories.items():
    json_object[key_for_documents].append({
        'clientId': k,
        'segmentName': v,
        key_for_documents: get_documents(dir_prefix=k)
    })

json = json.dumps(json_object, indent=4)
print(json)
```

## saving the JSON

```python
Path(f'{walk_root}_index/data/index.json').write_text(json, encoding='utf-8')
```
